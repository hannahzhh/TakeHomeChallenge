```
%sql

SELECT
  year(date) as year, 
  sum(TRU) as SUM_TRU,
  sum(DAU) as SUM_DAU,
  sum(NRU) as SUM_NRU,
  sum(Revenue) as sum_revenue,
  sum(Payer) as sum_payer
FROM topline_metrics
GROUP BY year(date)
ORDER BY year;
```

```
%sql

SELECT
  TRUNC(date, 'MM') as month, 
  sum(TRU) as SUM_TRU,
  sum(DAU) as SUM_DAU,
  sum(NRU) as SUM_NRU,
  sum(Revenue) as sum_revenue,
  sum(Payer) as sum_payer
FROM topline_metrics
GROUP BY TRUNC(date, 'MM')
ORDER BY month;
```
